# 연간 좋아요 수

```python
dic = {}
sum = 0
num = 0
year = 2009
for i in df['favorites']:
    if int(df['year'][num]) == year:
        sum += i
        if num == 41121:
            dic[year] = sum
    else:
        dic[year] = sum
        sum = 0
        year += 1
    num += 1
    
```

```python
import matplotlib.pyplot as plt

myList = dic.items()
x, y = zip(*myList)

plt.plot(x, y)
plt.show
```

```python
df1 = df['mentions'].dropna(axis = 0)
```

```python
df1
```

```python
from wordcloud import WordCloud, STOPWORDS
from collections import Counter

dict_data = Counter(df1)

dict_data


```

```python
wordcloud = WordCloud(width = 400, height = 400,
                      background_color = 'black',
                      stopwords = set(STOPWORDS),
                      max_words = 15,
                      max_font_size = 300,
                     ).generate_from_frequencies(dict_data)

plt.figure(figsize = (5, 5), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout()
plt.show()
                     
                     
```
